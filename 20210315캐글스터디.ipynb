## Evaluation

### Scoring Metric

Submissions are evaluated using the Normalized Gini Coefficient.
제출은 Normalized Gini Coefficient를 사용하여 평가됩니다.

During scoring, observations are sorted from the largest to the smallest predictions. Predictions are only used for ordering observations; therefore, the relative magnitude of the predictions are not used during scoring. The scoring algorithm then compares the cumulative proportion of positive class observations to a theoretical uniform proportion.
scoring 중에는 관측치가 가장 큰 예측에서 가장 작은 예측으로 정렬됩니다. 예측은 관측치의 순서 지정에만 사용되므로, 점수 매기기 중에 예측의 상대적 크기가 사용되지 않습니다. 그런 다음 scoring algorithm은 양의 클래스 관측치의 누적 비율을 이론적으로 동일한 비율과 비교합니다.

The Gini Coefficient ranges from approximately 0 for random guessing, to approximately 0.5 for a perfect score. The theoretical maximum for the discrete calculation is (1 - frac_pos) / 2.
Gini Coefficient의 범위는 무작위 추측의 경우 약 0에서 만점의 경우 약 0.5입니다. 이산형 계산에 대한 이론적 최대값은 (1 - frac_pos) / 2입니다.

The Normalized Gini Coefficient adjusts the score by the theoretical maximum so that the maximum score is 1.
Normalized Gini Coefficient는 이론적 최대값을 기준으로 점수를 조정하여 최대 점수가 1이 되도록 합니다.

The code to calculate Normalized Gini Coefficient in a number of different languages can be found in this forum thread.
여러 언어로 Normalized Gini Coefficient를 계산하는 코드느 이 포럼 스레드에서 찾을 수 있습니다.

### Target Value Distribution

```python
mpl.rc('font', size=15)
plt.figure(figsize=(7, 6))

# Target Value Distribution
ax = sns.countplot(x='target', data=train)

total_size = len(train) # Total number of train data

# Display the target value ratio at the top of the bar
for patch in ax.patches:
    height = patch.get_height() # Shape height (Data Count)
    width = patch.get_width() # Shape width
    left_coord = patch.get_x() # The x-axis position of the left border of the shape
    percent = height/total_size*100 # Target Value Ratio
    
    # Set text in coordinates (x, y)
    ax.text(x=left_coord + width/2.0, #x-axis position
            y=height + 3000, # y-axis position
            s='{:1.1f}%'.format(percent), # Text
            ha='center') # Center Alignment

ax.set_title('Target Distribution')
```

```python
def get_row_col_idx(idx, num_cols):
    # Subplot Row Index
    row_idx = idx//num_cols
    # Subplot Column Index
    for i in range(num_cols):
        if idx%num_cols == i:
            col_idx = i
    return row_idx, col_idx
```

```python
def plot_target_ratio_by_features(cols, nrows, ncols):
    mpl.rc('font', size=9) # Set font size
    figure, axes = plt.subplots(nrows=nrows, ncols=ncols) # Create subplots
    figure.set_size_inches(12, 18) # Set Figure size
    plt.subplots_adjust(wspace=0.3, hspace=0.3) # Set margins between subplots

    for idx, col in enumerate(train[cols]): 
        row_idx, col_idx = get_row_col_idx(idx, ncols) # Subplot row, column location
        ax = axes[row_idx, col_idx] # Set the axis on which to plot the graph
        # Plot bar graph of target value 1 ratio by unique values on the ax axis
        sns.barplot(x=col, y='target', data=train, palette='Set2', ax=ax)
```

LightGBM 장단점

장점 1. 학습하는데 걸리는 시간이 적다.

장점 2. 메모리 사용량이 상대적으로 적은 편이다.

장점 3. 카테고리형 피처들의 자동 변환과 최적 분할

단점 1. 적은 데이터 세트에 적용할 경우 과적합 가능성이 크다. (일반적으로 적은 데이터 세트의 기준은 공식 문서에서 10,000건 이하로 정의)

LightGBM 특징

* LightGBM은 리프 중심 트리 분할(Leaf Wise) 방식을 사용한다. 기존의 트리 기반 알고리즘은 균형 트리 분할(Level Wise) 방식을 사용한다.

* 기존의 트리 기반 알고리즘인 균형 트리 분할은 최대한 균형이 잡힌 트리를 유지하면서 분할하기 때문에 트리의 깊이가 최소화될 수 있다. 반대로 균형을 맞추기 위한 시간이 필요하다는 단점이 존재한다.

* 하지만 LightGBM의 리프 중심 트리 분할은 트리의 균형을 맞추지 않고 최대 손실 값(Max data loss)을 가지는 리프 노트를 지속적으로 분할하면서 트리의 깊이가 깊어지고 비대칭적인 트리가 생성된다. 그런데 최대 손실 값을 가지는 리프 노드를 반복할 수록 결국은 균형 트리의 분할 방식보다 예측 오류 손실을 최소화할 수 있다.

* num_leaves 개수를 높이면 정확도가 높아지지만, 트리의 깊이가 깊어지고 모델이 복잡해져 결국은 과적합 발생 가능성이 커진다.

* min_data_in_leaf 는 과적합을 위한 중요 파라미터이다. num_leaves와 학습 데이터의 크기에 따라 달라지겠지만 일반적으로 큰 값을 설정하게 되면 트리가 깊어지는 것을 방지해준다.

* max_depth 는 깊이의 크기를 제한한다. 위 두개의 파라미터들과 결합해서 과적합을 개선해준다.

### Training three simple models

```python
X_train = df_train.drop(['id', 'target'], axis=1)
y_train = df_train['target']
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, f1_score, roc_auc_score
from time import time

def metrics(true, preds):
    accuracy = accuracy_score(true, preds)
    recall = recall_score(true, preds)
    f1score = f1_score(true, preds)
    cf = confusion_matrix(true, preds)
    print('accuracy: {}, recall: {}, f1-score: {}'.format(accuracy, recall, f1score))
    print('Confusion matrix', cf)

def gini(true, preds):
    res = 2* roc_auc_score(true, preds) - 1
    return res

X_tr, X_te, y_tr, y_te = train_test_split(X_train, y_train, random_state=42, test_size=0.2)

def test_clfs(clfs):
    for clf in clfs:
        print('------------------------------------------')
        start = time()
        clf = clf(random_state=42)
        clf.fit(X_tr, y_tr)
        y1_pred = clf.predict(X_te)
        y1_pred_prob = clf.predict_proba(X_te)
        print(str(clf), '\nresults:')
        metrics(y_te, y1_pred)
        print('gini score', gini(y_te, y1_pred_prob[:, 1]))
        end = time()
        print('Processing time', end-start,'s')

classifiers = [LGBMClassifier, XGBClassifier, RandomForestClassifier]
test_clfs(classifiers)
```

Without parameter search, We can see that LGBMClassifier and XGBoost have better gini scores than randomforest. However, Lightgbm is mush faster and thus it can be helpful in extensive hyperparameter searching. Therefore, we will continue with parameter tuning only for Lightgbm.

parameter search가 없으면, LightGBM 분류기와 XGBoost가 randomforest보다 더 나은 gini score를 가지고 있음을 알 수 있다. 그러나 LightGBM은 더 빠르므로 광범위한 hyperparameter searching에 도움이 될 수 있다. 따라서 LightGBM에 대해서만 parameter tuning을 계속할 것이다.

It should be also noted that, as we can see from the confusion matrix, all the positive cases (1s) are missclassified, however we still have a very good gini score while recall is 0. The reason is for hard-classification the probability threshold of the classifier is 0.5. However, in this competition, we do not care about the hard predicted labels, and the only thing matters is the relative probability of the points in comparsion with each other. For example, we expect a higher probability for someone who is going to claim than someone who is not. Therefore, gini score is used in this competition and we will fine tune our parameters to maximize this score.

또한 confusion matrix에서 볼 수 있듯이 모든 positive case는 재현율(recall)이 0인 동안 여전히 매우 좋은 gini score를 가지고 있다. 그 이유는 하드 분류(hard-classification)의 경우 분류기의 확률 임계 값(probaility threshold)이 0.5이기 때문이다. 그러나 이 competition에서 우리는 hard predicted labels에 관심이 없으며, 중요한 것은 서로 비교되는 포인트의 상대 확률(relative probability of the points이다. 예를 들어, 우리는 주장하지 않는 사람보다 주장하려는 사람에게 더 높은 확률을 기대한다. 따라서 이 대회에서는 gini score가 사용되며 이 score를 최대화하기 위해 parameter를 fine tune할 것이다.

### Undersampling and Oversampling

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

sm_methods = [SMOTE,RandomUnderSampler,RandomOverSampler]
clf = LGBMClassifier(random_state=42)
for sm in sm_methods:
    print('------------------------------------------')
    sm = sm(random_state=42)
    start = time()
    X_train_resampled, y_train_resampled = sm.fit_resample(X_tr, y_tr)
    clf.fit(X_train_resampled, y_train_resampled)
    y_pred = clf.predict(X_te)
    y_pred_prob = clf.predict_proba(X_te)
    print(str(sm), 'results:')
    print('training resampled shape', X_train_resampled.shape)
    print('value counts in each class', y_train_resampled.value_counts())
    metrics(y_te, y_pred)
    gini_score = gini(y_te, y_pred_prob[:, 1])
    print('Normalized gini score', gini_score)
    end = time()
    print('Processing time', end-start,'s')
```
